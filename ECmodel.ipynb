# load data


```python
import pandas as pd
df=pd.read_csv('/content/global-data-on-sustainable-energy (1).csv')
df
```

# **Data preparation**


### **Data seperation as X and y**

```python
y=df['Primary energy consumption per capita (kWh/person)']
y
```

```python
X=df.drop('Primary energy consumption per capita (kWh/person)',axis=1)
X
```

## Data splitting

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train , y_test=train_test_split(X,y, test_size=0.2 , random_state=100)
```

```python
X_train
```

```python
X_test
```

# Model building

## Linear Regression

### Training the model

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Convert string columns in X_train to numerical using one-hot encoding
X_train = pd.get_dummies(X_train)

# Remove rows with any NaN values in X_train
X_train = X_train.dropna()

# If y_train contains categorical data, you may need to encode it as well
# y_train = pd.get_dummies(y_train)

# Align y_train with X_train after dropping rows
y_train = y_train[X_train.index]  # Keep only the y_train values corresponding to remaining X_train rows

lr = LinearRegression()
lr.fit(X_train, y_train)  # Now the fit should work without the error
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# ... (Your existing code from ipython-input-17-81967f6de6ea)

# Apply the SAME one-hot encoding to X_test, ensuring consistent features
X_test = pd.get_dummies(X_test)

# Align the columns of X_test with X_train (handling missing/extra columns)
# Use 'inner' join to keep only columns present in BOTH DataFrames
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# Impute remaining NaNs with a suitable strategy (e.g., mean)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Replace 'mean' with other strategies if needed
X_test = pd.DataFrame(imputer.fit_transform(X_test), columns=X_test.columns)

# Check for remaining NaNs after imputation (optional, for debugging)
print(X_test.isnull().sum().sum())

y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)  # Now the prediction should work
```

## Random forest

### Train modeling

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

Applying model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

Evaluate model **performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```
